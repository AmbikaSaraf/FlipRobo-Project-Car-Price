```python
for i in range(100):
    driver.get(link[i])
    time.sleep(2)
    #fetching car name
    try:
        car_t=driver.find_element_by_xpath("//div[@class='_35xN1']")
        car.append(car_t.text)
    except NoSuchElementException:
        car.append(np.nan)
    # fetching price
    try:
        price_t=driver.find_element_by_xpath("//div[@class='_3FkyT']")
        price.append(price_t.text)
    except NoSuchElementException:
        price.append(np.nan)
    # scrapig for number of owners
    try:
        own_t=driver.find_element_by_xpath("//div[@class='_1l939']/div[1]/div[2]/div[2]")
        owner.append(own_t.text)
    except NoSuchElementException:
        owner.append(np.nan)
    # scraping for distance covered
    try:
        dist_t=driver.find_element_by_xpath("//div[@class='aOxkz']/div[2]/div")
        dist.append(dist_t.text)
    except NoSuchElementException:
        dist.append(np.nan)
    # scraping for fuel type
    try:
        fuel_t=driver.find_element_by_xpath("//div[@class='aOxkz']/div[1]/div")
        fuel.append(fuel_t.text)
    except NoSuchElementException:
        fuel.append(np.nan)
    # scraping for transmission
    try:
        trans_t=driver.find_element_by_xpath("//div[@class='aOxkz']/div[1]/div")
        trans.append(trans_t.text)
    except NoSuchElementException:
        trans.append(np.nan)
    # scraping for year of purchase
    try:
        model_t=driver.find_element_by_xpath("//div[@class='_3tLee']")
        model.append(model_t.text)
    except NoSuchElementException:
        model.append(np.nan)
    
    website.append("olx.in")
```

#### Cars24.com

```python
driver.get("https://www.cars24.com/buy-used-car?sort=P&storeCityId=2&pinId=110001")
time.sleep(10)
```

```python
# Scraping for car name
car_t=driver.find_elements_by_xpath("//h2[@class='_3FpCg']")
car_name=[]
for i in car_t:
    car_name.append(i.text.replace("\n"," "))
for i in range(len(car_name)):
    st=car_name[i]
    car.append(st[4:])
    year.append(st[:4]) # year of purchase

# Scraping for price
price_t=driver.find_elements_by_xpath("//div[@class='_7udZZ']")
for i in price_t:
    price.append(i.text.replace("\n"," "))
    
# Scraping for number of owners
own_t=driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[2]")
for i in own_t:
    owner.append(i.text.replace("\n"," "))

# Scraping for distance covered
dist_t=driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[1]")
for i in dist_t:
    dist.append(i.text.replace("\n"," "))

# Scraping for fuel type
fuel_t=driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[3]")
for i in fuel_t:
    fuel.append(i.text.replace("\n"," "))

# Scraping for model name
model_t=driver.find_elements_by_xpath("//div[@class='_1jpRU']/p")
for i in model_t:
    model.append(i.text.replace("\n"," "))

# Transmission
try:
    trans_t=driver.find_elements_by_xpath("//div[@class='_1jpRU']/p/span")
except NoSuchElementException:
    trans_t.append(np.nan)
for i in trans_t:
        trans.append(i.text.replace("\n"," "))

# Product link
link_t=driver.find_elements_by_xpath("//a[@class='_9Ue0B']")
for i in link_t:
    link.append(i.get_attribute("href"))
```

```python
for i in car:
    website.append("cars24.com")
```

### Creating DataFrame

```python
df=pd.DataFrame()
df["CarName"]=car
df["Owner"]=owner
df["YearofPurchase"]=year
df["Transmission"]=trans
df["Model"]=model
df["Distance"]=dist
df["Fuel"]=fuel
df["Price"]=price
df["website"]=website
df["Product_url"]=link
```

```python
df.to_csv('data.csv')
```
